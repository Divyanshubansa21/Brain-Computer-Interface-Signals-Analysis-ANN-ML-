```python
# for data loading
import os
import pickle

import numpy as np
import pandas as pd

print("Libraries Imported Successfully ✅")
```

```python

# dataset path

dataset_path = "../dataset/data_preprocessed_python"

print("Dataset Path:")
print(dataset_path)
```

```python

#list of all dataset files

files = os.listdir(dataset_path)

print("Files in Dataset:")
print(files)
```

```python
print("Total Subjects:", len(files))
```

```python
# Load First Subject

file_path = os.path.join(dataset_path, "s01.dat")

print(file_path)
```

```python
# Open the First Subject File

with open(file_path, "rb") as file:
    subject_data = pickle.load(file, encoding="latin1")

print("Subject Loaded Successfully ✅")
```

```python
# Check Data Type

type(subject_data)
```

```python
# Display Keys

subject_data.keys()
```

```python
# Shape of Labels

print(subject_data["labels"].shape)
```

```python
# Shape of EEG Data

print(subject_data["data"].shape)
```

```python
# Display the keys present in the dataset

print(subject_data.keys())
```

```python
# Display Labels

labels = subject_data["labels"]

print(labels)
```

```python
# Convert labels into DataFrame

label_df = pd.DataFrame(
    labels,
    columns=["Valence", "Arousal", "Dominance", "Liking"]
)

label_df.head()
```

```python
label_df.describe()
```

```python
import matplotlib.pyplot as plt

label_df.hist(figsize=(10,6))

plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,6))

sns.heatmap(label_df.corr(),
            annot=True,
            cmap="coolwarm")

plt.show()
```

```python
eeg_data = subject_data["data"]

print(eeg_data.shape)
```

```python
eeg_only = eeg_data[:,0:32,:]

print(eeg_only.shape)
```

```python
first_trial = eeg_only[0]

print(first_trial.shape)
```

```python
plt.figure(figsize=(15,4))

plt.plot(first_trial[0])

plt.title("EEG Signal - Channel 1")

plt.xlabel("Samples")

plt.ylabel("Amplitude")

plt.show()
```

```python
plt.figure(figsize=(15,4))

plt.plot(first_trial[10])

plt.title("EEG Signal - Channel 11")

plt.show()
```

```python
print(first_trial.max())

print(first_trial.min())
```

```python
print(first_trial.mean())

print(first_trial.std())
```

```python
# Separate EEG and Labels

X = subject_data["data"][:, :32, :]
y = subject_data["labels"]

print("EEG Shape :", X.shape)
print("Labels Shape :", y.shape)
```

```python
# Convert Labels to DataFrame

label_df = pd.DataFrame(
    y,
    columns=[
        "Valence",
        "Arousal",
        "Dominance",
        "Liking"
    ]
)

label_df.head()
```

```python
label_df.describe()
```

```python
import matplotlib.pyplot as plt

label_df.hist(figsize=(10,8))

plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,6))

sns.heatmap(
    label_df.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.show()
```
